```python
import seaborn as sns
import matplotlib.pyplot as plt
# to install seaborn library please follow the following instruction
# pip install seaborn
# conda install seaborn
import numpy as np
```

```python
tips=sns.load_dataset('tips') # seaborn has three default dataset - tips,iris,flight
# load_dataset() will load the dataset and the data type will be dataframe.
```

```python
tips.head()
```

```python
fig=plt.figure(figsize=(10,8))
sns.distplot(tips['total_bill'],kde=False,bins=20,color='blue')
```

```python
sns.rugplot(tips['total_bill'],color='red')
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(tips['total_bill'],y=tips['tip'],kind='resid')
```
