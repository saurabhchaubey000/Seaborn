```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
fig=plt.figure(figsize=(10,8))
```

```python
sns.distplot(tips['total_bill'],bins=15,hist=True,kde=True,rug=True,color='r')
```

```python
sns.rugplot(tips['total_bill'],tips['tip'],height=.6,axis='y')
```

```python
sns.kdeplot(tips['total_bill'],kernel='epa')
```

```python
sns.jointplot(tips['total_bill'],y=tips['tip'],kind='hex')
```

```python
#categorical graph
```

```python
sns.barplot(y="total_bill",x='smoker',data=tips,hue='sex',estimator=np.min,color='r')
```

```python
sns.boxplot(y='total_bill',x='day',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```
